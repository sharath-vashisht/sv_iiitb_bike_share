## Inference from data:
- We can see total 730 rows and 16 columns all having non- null value, so cleanup of data is not needed.
- "instant" field can be dropped from the table at is is only gives the row number.
- For dteday is a timestamp field. We already have yr and month data.
- casual and registered can alsob ed dropped as we can see that cnt is sum of both casual and registered
- Eventhough below fields are integer in nature, they are categorical variables.
  - Year
  - season
  - month
  - holiday
  - weekday
  - workingday
  - weathersit
- Out of which certain data are already binary in nature so dummy variables need to be created for below fields
  - year
  - season
  - weekday 
  - Month
  - weathersit

```python
# Drop instant and dteday columns
df=df.drop(["instant","dteday","casual", "registered"], axis=1)
df.head()
```

```python
#Enter proper values for field from data dictionary. This will help in creating proper names for the dummy variable.
#	- yr : year (0: 2018, 1:2019)
df["yr"] = df["yr"].map({0:"2018", 1:"2019"})

#Month
months = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}

df["mnth"] = df["mnth"].map(months)

#In data dictionary definition of weekday is not given. However, when googled about the weekday of 01-01-2028 it was found that it was on monday and corresponding weekday mapped was 6 in data set. 
#So 6 will be monday, 0 will be tuesday and hence creating columns accordingly
days = {
    0: 'Tue',
    1: 'Wed',
    2: 'Thu',
    3: 'Fri',
    4: 'Sat',
    5: 'Sun',
    6: 'Mon'
}

df["weekday"] = df["weekday"].map(days)

#Season 1:spring, 2:summer, 3:fall, 4:winter
df["season"] = df["season"].map({1:"spring", 2: "summer", 3: "fall" , 4:"winter"})

# weathersit : 
		#- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		#- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		#- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		#- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
df["weathersit"] = df["weathersit"].map({1:"Clear", 2:"Mist", 3: "Light Snow and Rain", 4:"Heavyrain"} )
```

## Visualization Inference: 
- From the above data we can see there is some kind of linear relationship between count vs temp, count vs atemp, count vs cacual user, count vs registered user.
- Count of users are high when during fall, summer and winter. However during spring there is a drop around 25%
- People are using bike more during clear weather than in light snow.
- Business is showing positive trend from 2018 to 2019 showing the popularity of the service.
- There is not much impact whether it is a working day or holiday.

```python
# Creation of dummies for year, month, season and weatherlist
year = pd.get_dummies(df["yr"], drop_first=True)
df = pd.concat([df, year], axis=1)

weekday = pd.get_dummies(df["weekday"], drop_first=True)
df = pd.concat([df, weekday], axis=1)

mnth = pd.get_dummies(df["mnth"], drop_first=True)
df = pd.concat([df, mnth], axis=1)

season = pd.get_dummies(df["season"], drop_first=True)
df = pd.concat([df, season], axis=1)

weathersit = pd.get_dummies(df["weathersit"], drop_first=True)
df = pd.concat([df, weathersit], axis=1)
```
